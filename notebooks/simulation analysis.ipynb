```python
# Load data
df = pd.read_excel('../data/raw/графики.xlsx', sheet_name='Лист1')
df['GigaChat3'] = df['GigaChat3'].ffill()

# === Select cluster ===
cluster_name = 'Cluster 0'  # change to 'Cluster 1' if needed
cluster_data = df[df['GigaChat3'].str.contains(cluster_name, na=False)]

metrics_raw = [
    'mean_similarity',
    'mean_mae_35',
    'mean_similarity facets',
    'mean_similarity traits'
]
metrics_display = [
    'Similarity on Answers %',
    'MAE on Traits & Facets',
    'Similarity on Facets %',
    'Similarity on Traits %'
]

non_opt = cluster_data[cluster_data['Prompt'] == 'non opt'].iloc[0][metrics_raw].values.astype(float)
opt = cluster_data[cluster_data['Prompt'] == 'opt'].iloc[0][metrics_raw].values.astype(float)

# Convert similarity metrics to %
for i in [0, 2, 3]:
    non_opt[i] *= 100
    opt[i] *= 100

# DataFrame for plotting
plot_df = pd.DataFrame({
    'Metric': metrics_display * 2,
    'Value': list(non_opt) + list(opt),
    'Type': ['Non-Optimized'] * 4 + ['Optimized'] * 4
})

# === Plot ===
fig, ax = plt.subplots(figsize=(10.5, 6.5))

sns.barplot(
    data=plot_df,
    x='Metric',
    y='Value',
    hue='Type',
    palette='muted',
    ax=ax
)

# Add value labels on top of bars
for container in ax.containers:
    ax.bar_label(
        container,
        fmt='%.2f',
        fontsize=9,
        padding=3
    )

# Highlight differences (Δ) more clearly
for i in range(len(metrics_display)):
    delta = opt[i] - non_opt[i]
    y_pos = max(non_opt[i], opt[i]) * 1.08

    ax.text(
        i,
        y_pos,
        f'Δ = {delta:+.2f}',
        ha='center',
        va='bottom',
        fontsize=11,
        fontweight='bold',
        color='black'
    )

# Legend placement (top-right, not overlapping bars)

ax.legend(
    title='',
    loc='upper left',

    frameon=True
)

plt.tight_layout(rect=[0, 0, 0.85, 1])

ax.spines['top'].set_visible(False)
plt.xticks(rotation=30, ha='right')
plt.ylabel('Score')

ax.set_facecolor("#FFE4C4")
fig.patch.set_facecolor("#FFE4C4")

plt.show()
```
